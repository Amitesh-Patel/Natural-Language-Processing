```python
import re
```

```python
text = "Hi this is mNine . whats up how going is everything ok"
```

```python
'this' in text
```

```python
my_match = re.search('this',text)
```

```python
my_match.span()
```

```python
 match = re.findall('is',text)
```

```python
len(match)
```

```python
for match in re.finditer('is',text):
    print(match.span())
```
